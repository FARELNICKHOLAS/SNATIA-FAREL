```python
import pandas as pd

data = pd.read_csv('datastemming.csv')

data.head(10)
```

```python
data = data.dropna()
```

```python
def tokenisasi(text):
    tokens = text.split()
    return tokens

data['Tokenisasi'] = data['stemmed_data'].apply(tokenisasi)
data.head(5)
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
factory = StopWordRemoverFactory()
stop_words = factory.get_stop_words()
print(stop_words)
```

```python
def remove_stopwords(text):
  return [word for word in text if word not in stop_words]

data['data_done'] = data['Tokenisasi'].apply(lambda x: remove_stopwords(x))

data.head(10)
```

```python
import matplotlib.pyplot as plt

sentimen_count = data['sentimen'].value_counts()

plt.bar(sentimen_count.index, sentimen_count.values, color=['red', 'green'])
plt.xlabel('Sentimen')
plt.ylabel('Jumlah')
plt.title('Perbandingan Jumlah Sentimen Negatif dan Positif')
plt.show()
```

```python
score_counts = data['sentimen'].value_counts()

print(score_counts)
```

```python
d1 = data[data['sentimen'] == 'negatif'].sample(4997, replace = False, random_state = 0)
d2 = data[data['sentimen'] == 'positif'].sample(4997, replace = False, random_state = 0)
df2 = pd.concat([d1, d2])
```

```python
score_counts = df2['sentimen'].value_counts()

print(score_counts)
```
